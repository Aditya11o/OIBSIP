### Project Overview
Customer segmentation is the process of partitioning a customer base into sub-groups of individuals who share similar characteristics (such as demographics, spending score, income limits, or product affinities). 

In this project, we analyze a dataset of mall customers using unsupervised machine learning. By utilizing **K-Means Clustering**, we segment customers based on their **Annual Income** and **Spending Score**. This segmentation allows retail and e-commerce companies to design targeted marketing strategies, adjust pricing, and optimize customer satisfaction.

## 6. K-Means Clustering & The Elbow Method
To segment customers, we extract their **Annual Income (k$)** and **Spending Score (1-100)** features. We then run K-Means iteratively from $K=1$ to $K=10$ to find the optimal number of clusters using the **Within-Cluster Sum of Squares (WCSS)**.


```python
# Extract Income and Spending Score features
X = df.iloc[:, [3, 4]].values

# Calculate WCSS for K=1 to K=10
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Curve
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', color='#e53e3e', linewidth=2, markersize=8)
plt.title('The Elbow Method for Optimal K', pad=20, weight='bold', fontsize=15)
plt.xlabel('Number of Clusters (K)', labelpad=10)
plt.ylabel('WCSS (Within-Cluster Sum of Squares)', labelpad=10)

# Annotate the elbow point
plt.annotate('Elbow Point (K=5)', xy=(5, wcss[4]), xytext=(6, wcss[4] + 30000),
             arrowprops=dict(facecolor='#2d3748', shrink=0.08, width=1.5, headwidth=6),
             horizontalalignment='left', weight='bold', color='#2d3748')

plt.xticks(range(1, 11))
plt.tight_layout()
plt.savefig('../Visualizations/elbow_method.png', dpi=300)
plt.show()
```

## 7. Fitting K-Means with K=5
As seen in the Elbow plot, the drop in WCSS slows down significantly after $K=5$. Therefore, $K=5$ is selected as the optimal number of segments. We fit the KMeans model and assign segment labels to the dataset.


```python
# Instantiate K-Means with K=5
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42, n_init=10)

# Predict clusters
y_kmeans = kmeans.fit_predict(X)

# Assign labels back to dataframe
df['Cluster'] = y_kmeans
df.head()
```

## 8. Visualizing the Customer Segments
We map the K-Means cluster indexes to clear business descriptive labels and plot them with unique colors, marking the centroids clearly.


```python
plt.figure(figsize=(12, 8))

# Descriptive labels mapped to cluster indexes
cluster_labels = {
    0: 'Standard Customers (Mid Income, Mid Spend)',
    1: 'VIP Target (High Income, High Spend)',
    2: 'Spendthrifts (Low Income, High Spend)',
    3: 'Careful Spenders (High Income, Low Spend)',
    4: 'Sensible Spend (Low Income, Low Spend)'
}

colors_list = ['#4299e1', '#48bb78', '#9f7aea', '#ed64a6', '#ecc94b']

# Plot each cluster
for i in range(5):
    plt.scatter(X[y_kmeans == i, 0], X[y_kmeans == i, 1], s=85, 
                c=colors_list[i], label=cluster_labels[i], alpha=0.85, edgecolors='w', linewidths=0.5)

# Plot centroids
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], s=250, c='#1a202c', marker='X', 
            edgecolors='w', linewidths=1.5, label='Centroids')

plt.title('Customer Segments using K-Means Clustering', pad=25, weight='bold', fontsize=16)
plt.xlabel('Annual Income (k$)', labelpad=12)
plt.ylabel('Spending Score (1-100)', labelpad=12)
plt.legend(title='Customer Segments', frameon=True, facecolor='w', edgecolor='#cbd5e0', loc='upper right')
plt.tight_layout()
plt.savefig('../Visualizations/customer_segments.png', dpi=300)
plt.show()
```

## 9. Customer Segment Profiling
We calculate aggregate statistics (count, mean age, mean income, and mean spending score) for each group to construct clear segment personas.


```python
# Aggregate cluster statistics
cluster_profiles = df.groupby('Cluster').agg(
    Count=('CustomerID', 'count'),
    Avg_Age=('Age', 'mean'),
    Avg_Income=('Annual Income (k$)', 'mean'),
    Avg_Spending_Score=('Spending Score (1-100)', 'mean')
).reset_index()

# Map labels to profile dataframe
cluster_profiles['Segment Name'] = cluster_profiles['Cluster'].map(cluster_labels)
cluster_profiles = cluster_profiles[['Cluster', 'Segment Name', 'Count', 'Avg_Age', 'Avg_Income', 'Avg_Spending_Score']]
cluster_profiles
```

```python
# Demographic Gender breakdown inside segments
gender_profiles = df.groupby(['Cluster', 'Gender']).size().unstack(fill_value=0)
gender_profiles.index = gender_profiles.index.map(cluster_labels)
gender_profiles
```

## 10. Key Business Insights & Recommendations

Based on the 5 distinct customer personas discovered:

1. **VIP Target (High Income, High Spend)**:
   - *Persona*: Young adults (average age 32.7) with high purchasing power and high spending score.
   - *Strategy*: Retain and delight this premium segment with early access to new collections, exclusive loyalty programs, personal shoppers, and tailored high-end rewards.
2. **Careful Spenders (High Income, Low Spend)**:
   - *Persona*: Mature buyers (average age 41.1) with high income but conservative spending.
   - *Strategy*: Re-engage them with targeted value propositions, premium bundles, and direct marketing emphasizing high product quality, warranty, and utilities.
3. **Standard Customers (Mid Income, Mid Spend)**:
   - *Persona*: Middle-aged customers (average age 42.7) showing moderate spending.
   - *Strategy*: Maintain their loyalty using standard seasonal newsletters, discount coupons, and point-of-sale cashbacks.
4. **Spendthrifts (Low Income, High Spend)**:
   - *Persona*: Youngest customer group (average age 25.3) who love to spend despite lower income limits.
   - *Strategy*: Drive volume using trendy, budget-friendly product drops, peer marketing, and app-only deals.
5. **Sensible Spend (Low Income, Low Spend)**:
   - *Persona*: Oldest cohort (average age 45.2) with low income and low spending.
   - *Strategy*: Target them with deep discount pricing, value-focused clearance sales, and basic necessities.
